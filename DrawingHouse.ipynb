```python
import turtle
t = turtle.Turtle()
t.speed(10)
turtle.bgcolor('SkyBlue');
```

```python
### Vẽ ngôi nhà
t.penup()
t.goto(-300, -300)
t.pendown()
t.fillcolor('blue')
t.begin_fill()
for i in range (2):
    t.fd(120)
    t.lt(90)
    t.fd(150)
    t.lt(90)
t.end_fill()

t.penup()
t.goto(-300, -150)
t.pendown()
t.lt(60)
t.fillcolor('pink')
t.begin_fill()
t.fd(120)
t.rt(120)
t.fd(120)
t.end_fill()

t.fillcolor('orange')
t.begin_fill()
for i in range (4):
    t.lt(90)
    t.fd(120)
t.end_fill()

t.fillcolor('yellow')
t.begin_fill()
t.rt(30)
t.fd(150)
t.lt(120)
t.fd(120)
t.lt(60)
t.fd(150)
t.end_fill()

t.penup()
t.goto(-270, -300)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.fd(75)
t.rt(90)
t.fd(60)
t.rt(90)
t.fd(75)
t.end_fill()

t.penup()
t.goto(-150, -165)
t.pendown()
t.fillcolor('brown')
t.begin_fill()
t.fd(60)
t.lt(120)
t.fd(60)
t.lt(60)
t.fd(60)
t.lt(120)
t.fd(60)
t.end_fill()
```

```python
### Vẽ cây
# Gốc cây
t.penup()
t.goto(200, -350)
t.pendown()
t.fillcolor('brown')
t.begin_fill()
t.lt(150)
for i in range (2):
    t.fd(40)
    t.lt(90)
    t.fd(60)
    t.lt(90)
t.end_fill()

# Cành cây 1
t.penup()
t.goto(220, -290)
t.pendown()
t.fillcolor('lightgreen')
t.begin_fill()
t.fd(80)
t.lt(135)
t.fd(113)
t.lt(90)
t.fd(113)
t.lt(135)
t.fd(80)

# Cành cây 2
t.penup()
t.goto(220, -210)
t.pendown()
t.fd(80)
t.lt(135)
t.fd(113)
t.lt(90)
t.fd(113)
t.lt(135)
t.fd(80)

# Cành cây 3
t.penup()
t.goto(220, -130)
t.pendown()
t.fd(80)
t.lt(135)
t.fd(113)
t.lt(90)
t.fd(113)
t.lt(135)
t.fd(80)
t.end_fill()
```

```python
### Vẽ mặt trời
# Tia nắng
t.penup()
t.goto(150, 200)
t.pendown()
for i in range (2):
    t.fd(100)
    t.bk(100)
    t.lt(180)
    t.fd(100)
    t.bk(100)
    t.rt(45)
    t.fd(80)
    t.bk(80)
    t.lt(180)
    t.fd(80)
    t.bk(80)
    t.rt(45)

# Mặt trời
t.goto(150, 250)
t.fillcolor('yellow')
t.begin_fill()
t.circle(50)

t.end_fill()
```

```python
turtle.done()
```
